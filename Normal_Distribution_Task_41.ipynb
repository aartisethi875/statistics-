# Which feature is the best predictor of the wine class in the wine dataset? (Using pdf plot)

````python

# load the wine dataset
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target
sns.set_style('whitegrid')
sns.set_palette('colorblind')



# plot PDF for each feature for each class
for i, feature in enumerate(data.feature_names):
    fig, ax = plt.subplots(figsize=(8, 6))
    for target in df['target'].unique():
        sns.histplot(df[feature][df['target']==target],label=f'Class {target}', kde=True)
    ax.set_title(f'{feature} Distribution')
    ax.set_xlabel(feature)
    ax.set_ylabel('Density')
    ax.legend()

plt.tight_layout()
plt.show()
````

````python
# Define colors for each class
colors = ['#34495E', '#6C3483', '#EC7063']

# Create a figure with subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs = axs.ravel()

# Loop through each pair of features
for i, feature1 in enumerate(["proline", "flavanoids", "color_intensity", "od280/od315_of_diluted_wines", "alcohol"]):
    for j, feature2 in enumerate(["proline", "flavanoids", "color_intensity", "od280/od315_of_diluted_wines", "alcohol"]):
        if j > i:
            axs[i].set_xlabel(feature1)
            axs[i].set_ylabel(feature2)
            for target, color in enumerate(colors):
                axs[i].scatter(df[df.target==target][feature1], df[df.target==target][feature2], 
                                  alpha=0.5, label=f'class {target}', c=color)
            axs[i].legend()
            
plt.tight_layout()
plt.show()


````
